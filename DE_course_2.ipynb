```
okt = Okt()
print(okt.morphs(text))
print(okt.morphs(text, stem=True))
print(okt.morphs(text, norm=True))
print("*"*100)

print(okt.nouns(text))
print(okt.phrases(text))
print("*"*100)

print(okt.pos(text))
print(okt.pos(text, join=True))   # 형태소 + 품사 -> 리스트로
```
